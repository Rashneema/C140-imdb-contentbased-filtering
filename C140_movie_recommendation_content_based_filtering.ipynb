**C139 STARTS HERE**

Formula to calculate weighted movie rating avg

(v/(v+m))*R)+((m/(v+m))*C)




```python
c=df2["vote_average"].mean()
c
```

```python
m=df2['vote_count'].quantile(0.9)
m
```

```python
q_movies = df2[df2['vote_count']>=m]
q_movies
```

```python
q_movies.shape
```

```python
#function to calculate the metric score
def weighted_rating(x,m=m,c=c): #x is each row
  v=x["vote_count"]
  r=x['vote_average']
  score = (v/(v+m)*r) + (m/(m+v)*c)
  return score
```

```python
#Applying function on each row
q_movies['score']=q_movies.apply(weighted_rating,axis=1)
```

```python
q_movies.head()
```

```python
#sorting the dataframe by score and displaying only selected columns 
q_movies = q_movies.sort_values('score',ascending=False)
q_movies[['title_x','vote_count','vote_average','score']].head(10)
```

**CONTENT BASED FILTERING**

```python
df2[['title_x', 'cast', 'crew', 'keywords', 'genres']].head(3)
```

```python
from ast import literal_eval

features = ['cast', 'crew', 'keywords', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(literal_eval)

df2.dtypes
```

```python
df2['crew'][0]
```

```python
import numpy as np
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan

df2['director'] = df2['crew'].apply(get_director)
```

```python
df2.head(1)
```

```python
df2['cast'][0]
```

```python
def get_list(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        return names
    return []

features = ['cast', 'keywords', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(get_list)
```

```python
df2.head(1)
```

```python
df2['cast'][0]
```

```python
df2[['title_x', 'cast', 'director', 'keywords', 'genres']].head(3)
```

```python
def clean_data(x):
    if isinstance(x, list): #checkig if it is a list
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        if isinstance(x, str): #else checking if it is a str
            return str.lower(x.replace(" ", ""))
        else:
            return ''

features = ['cast', 'keywords', 'director', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(clean_data)
```

```python
df2.head(1)
```

```python
def create_soup(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])
df2['soup'] = df2.apply(create_soup, axis=1) 
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df2['soup'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
```

```python
df2 = df2.reset_index()
indices = pd.Series(df2.index, index=df2['title_x'])
```

```python
df2.head()
```

```python
def get_recommendations(title, cosine_sim):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]
    return df2['title_x'].iloc[movie_indices]
```

```python
get_recommendations('Fight Club', cosine_sim2)
```

```python
get_recommendations('The Shawshank Redemption', cosine_sim2)
```
